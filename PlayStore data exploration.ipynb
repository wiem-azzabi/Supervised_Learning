# exploration et analyse des données

```python
# L'App Store et le Play Store accordent plus d'importance que jamais aux notes(Rating) et aux avis(Reviews). 
# Les applications avec des notes et des avis plus élevés sont bien classées dans la recherche.
# Si une application est bien classée, il y a plus de chances qu'elle soit trouvée et téléchargée.
# Plus il y aura de téléchargements, plus il sera maintenu en haut du classement.
```

```python
data["Type"] = data["Type"].astype('category')
data["Category"] = data["Category"].astype('category')
data["Android Ver"]= data["Android Ver"].astype('category')
```

```python
data["Type_cat"] = data["Type"].cat.codes
data["Category_cat"] = data["Category"].cat.codes
data["Android Ver_cat"] = data["Android Ver"].cat.codes

```

```python
data["Current Ver"]= data["Current Ver"].astype('category')
data["Current Ver_cat"]= data["Current Ver"].cat.codes
```

```python
# data.drop()
```

```python
data["Reviews"].nunique()
```

```python
#Transformer les Reviews en type int
# Transformer les dates en type date 
```

```python
# data[data['Reviews'].str.contains('M')]
# data.drop(data.index[10472],inplace=True)
data["Reviews"]=data["Reviews"].astype(int)
```

```python
# transformer "size" en type numerique 1M=10^6 , 1K=10^3
```

```python
data["Size"].unique()
```

```python
data['Size'].replace('Varies with device', np.nan, inplace = True ) 
# data2[data2['Size'].str.contains('M')].replace("M", "",regex=True)
data.Size = (data.Size.replace(r'[kM]+$', '', regex=True).astype(float) * \
             data.Size.str.extract(r'[\d\.]+([KM]+)', expand=False)
            .fillna(1)
            .replace(['k','M'], [10**3, 10**6]).astype(int))
data['Size'].fillna(data.groupby('Category')['Size'].transform('mean'),inplace = True)
# fill "Varies with device" with mean of size in each category
```
